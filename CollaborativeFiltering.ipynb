# **Preprocessing**

```python
import pandas as pd
```

```python
df=pd.read_csv("viewer_actions (1).csv")
df
```

```python
df.dropna(inplace=True)
df
```

```python
df.drop(["content_type"],axis=1,inplace=True)
```

```python
df
```

# **Video Recommendation Engine**

```python
df_v = df[df['action_type'] <= 5]
df_v
```

```python
dfv_prep=pd.crosstab(df_v['viewer_id'],df_v['content_id']).astype(int)
dfv_prep
```

```python
content_similarity_df=dfv_prep.corr(method='pearson')
content_similarity_df.head(50)
```

```python
ratings_matrix = pd.DataFrame(content_similarity_df)
ratings_matrix.head(15)
```

```python
def get_recom(content_id):
  similar_score=content_similarity_df[content_id]
  similar_score=similar_score.sort_values(ascending=False)
  return similar_score
pd.DataFrame(get_recom(13)).drop(13).head(20)
```

```python
user=[(1639),(51),(49)]
similar_content=pd.DataFrame()
for content_id in user:
  similar_content=similar_content.append(get_recom(content_id),ignore_index=True)
similar_content.head()
pd.DataFrame(similar_content.sum().sort_values(ascending=False)).drop(user).head(20)
```

# **Audio Recommendation Engine**

```python
df_a=df[df['action_type'] >= 7]
df_a
```

```python
dfa_prep=pd.crosstab(df_a['viewer_id'],df_a['content_id']).astype(int)
dfa_prep
```

```python
content_similarity_dfa=dfa_prep.corr(method='pearson')
content_similarity_dfa.head(10)
```

```python
ratings_matrixa = pd.DataFrame(content_similarity_dfa)
ratings_matrixa.head(15)
```

```python
def get_recom(content_id):
  similar_score=content_similarity_dfa[content_id]
  similar_score=similar_score.sort_values(ascending=False)
  return similar_score
pd.DataFrame(get_recom(459)).drop(459).drop([459],axis=1).head(20)
```

```python
user=[(2101),(353),(475),(2127),(459)]
similar_content=pd.DataFrame()
for content_id in user:
  similar_content=similar_content.append(get_recom(content_id),ignore_index=True)
similar_content.head()
pd.DataFrame(similar_content.sum().sort_values(ascending=False)).drop(user).drop([0],axis=1).head(20)
```
